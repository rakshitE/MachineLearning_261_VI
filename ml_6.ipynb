```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# 1. Load a sample dataset (Iris)
iris = datasets.load_iris()
X = iris.data[:, :2]  # We'll only use the first two features for easy plotting
y = iris.target

# To keep it simple, let's do a binary classification (ignore the third class)
X = X[y != 2]
y = y[y != 2]

# 2. Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Create and train the SVM model
# 'linear' kernel is great for simple, separable data
model = SVC(kernel='linear', C=1.0)
model.fit(X_train, y_train)

# 4. Make predictions
predictions = model.predict(X_test)

# 5. Evaluate the results
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
