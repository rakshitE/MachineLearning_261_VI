```python
import numpy as np

class PCA:
    def __init__(self, k):
        self.k = k                  # number of components
        self.components = None      # eigenvectors
        self.mean = None

    def fit(self, X):
        # Step 1: Mean centering
        self.mean = np.mean(X, axis=0)
        X_centered = X - self.mean

        # Step 2: Covariance matrix
        cov_matrix = np.cov(X_centered, rowvar=False)

        # Step 3: Eigen decomposition
        eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

        # Step 4: Sort eigenvectors by eigenvalues (descending)
        sorted_indices = np.argsort(eigenvalues)[::-1]
        eigenvectors = eigenvectors[:, sorted_indices]

        # Step 5: Select top k eigenvectors
        self.components = eigenvectors[:, :self.k]

    def transform(self, X):
        # Project data
        X_centered = X - self.mean
        return np.dot(X_centered, self.components)

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

# Sample dataset (rows = samples, columns = features)
X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
])

# Reduce to 1 dimension
pca = PCA(k=1)
Z = pca.fit_transform(X)

print("Reduced Data:\n", Z)
```
